## Remote Work and Mental Health Project

## Introduction

Remote Work and Mental Health: A Critical Topic
With the rise of remote work, especially after the COVID-19 pandemic, understanding its impact on employee satisfaction and mental health has become essential. This project aims to explore these themes using a dataset titled "Impact of Remote Work on Mental Health".

Why is this project important for a Data Analyst?
Data wrangling, exploratory analysis, and visualization are core skills for any data analyst. This project highlights the key stages of:

Data Wrangling: Cleaning and transforming raw data for analysis.
Exploratory Data Analysis (EDA): Extracting meaningful insights from the dataset.
Data Visualization: Presenting findings through clear and impactful visuals.

Objectives of the Project:

Investigate the impact of stress, social isolation, and company support on remote work satisfaction.
Identify groups of employees most affected by remote work challenges.
Provide actionable recommendations to improve employee mental health and satisfaction.

Dataset Description
Source: This dataset contains information on 5000 employees working across various industries and regions.
Key Features:

Demographics:
Age, Gender, Region.
Professional Information:
Job_Role, Industry, Work_Location, Years_of_Experience.
Mental Health and Well-Being:
Stress_Level, Social_Isolation_Rating, Mental_Health_Condition.
Work Performance:
Hours_Worked_Per_Week, Number_of_Virtual_Meetings, Productivity_Change.
Satisfaction:
Work_Life_Balance_Rating, Satisfaction_with_Remote_Work, Company_Support_for_Remote_Work.
Each variable plays a significant role in assessing how remote work impacts employees' mental health and productivity.

```python
## Exploratory Data Analysis (EDA):
#Perform univariate, bivariate, and multivariate analysis.
#Investigate the relationship between satisfaction and other variables (e.g., stress, work location).
```

```python
# Age Distribution of Employees
# Objective: Understand the distribution of employees by age to identify patterns in the workforce.

plt.figure(figsize=(10, 6))
sns.histplot(remote_work['Age'], bins=15, kde=True, color='#eb984e')
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Hours Worked Per Week by Job Role
# Objective: Analyze the number of hours worked per week across different job roles to identify workload patterns.
plt.figure(figsize=(12,6))
sns.boxplot(
    x="Job_Role",
    y="Hours_Worked_Per_Week",
    data=remote_work,
    hue=None,  # vite l'avertissement
    color='#76d7c4'
)
plt.title("Hours Worked Per Week by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Hours Worked Per Week")
plt.xticks(rotation=45)
plt.show()

```

```python
# Gender Distribution of Employees
# Objective: Visualize the proportion of employees by gender.

remote_work["Gender"].value_counts().plot(kind="bar",color=["#f39c12","#ba4a00","#935116","#f4d03f"],edgecolor="black")
plt.ylim(1000, 1300)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Work Location Distribution
# Objective: Show the distribution of employees based on their work location (remote, hybrid, onsite).

remote_work['Work_Location'].value_counts().plot(kind='bar',color= ['#76d7c4','#2e86c1','#2ecc71'],edgecolor='black')
plt.ylim(1500, 1700)
plt.title("Work Location Distribution")
plt.xlabel("Work Location")
plt.ylabel("Count")
plt.show()
```

```python
# Satisfaction Levels by Work Location
# Objective: Compare employee satisfaction levels with remote work across different work locations.


satisfaction_counts = pd.crosstab(remote_work['Work_Location'], remote_work['Satisfaction_with_Remote_Work'], normalize='index')

satisfaction_counts.plot(kind='bar', stacked=True, color=['#ffadad', '#ffd6a5', '#caffbf'], edgecolor='black', linewidth=1.5)
plt.title("Satisfaction Levels by Work Location")
plt.xlabel("Work Location")
plt.ylabel("Proportion")
plt.legend(title="Satisfaction Level", bbox_to_anchor=(1, 1))
plt.xticks(rotation=0)
plt.show()

```

```python
# Mental Health Condition by Work Location
# Objective: Examine the distribution of mental health conditions among employees based on their work location.

sns.violinplot(
    x="Work_Location", 
    y="Mental_Health_Condition", 
    data=remote_work, 
    hue=None,  
    density_norm="count", 
    palette="viridis"
)
plt.title("Mental Health Condition by Work Location")
plt.xlabel("Work Location")
plt.ylabel("Mental Health Condition")
plt.show()
```

```python
# Number of Remote Workers by Age Group
# Objective: Analyze the age distribution of employees who work remotely to identify trends by age group.


# Create age groups
remote_work['Age_Group'] = pd.cut(
    remote_work['Age'],
    bins=[18, 25, 35, 45, 55, 65, 100],  # Ajuster les groupes d'ge
    labels=["18-24", "25-34", "35-44", "45-54", "55-64", "65+"],
    right=False
)

# Filter employees working remotely
remote_workers = remote_work[remote_work['Work_Location'] == 'Remote']

# Count remote workers by age group
remote_by_age_group = remote_workers['Age_Group'].value_counts().sort_index()

# Plot the bar chart
remote_by_age_group.plot(
    kind='bar',
    color=['#b2dfdb', '#80cbc4', '#4db6ac', '#26a69a', '#00897b', '#00695c'],  # Palette personnalise
    edgecolor='black'
)
plt.title("Number of Remote Workers by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Remote Workers")
plt.xticks(rotation=45) 
plt.show()
```

```python
# Convert Productivity_Change to numerical values
productivity_map = {'Increase': 1, 'No Change': 0, 'Decrease': -1}
remote_work['Productivity_Change_Num'] = remote_work['Productivity_Change'].map(productivity_map)

# Check the unique stress levels
print(remote_work['Stress_Level'].unique())
```

```python
# Average productivity by job role and stress level
productivity_by_job_stress = remote_work.groupby(['Job_Role', 'Stress_Level'])['Productivity_Change_Num'].mean().unstack()

print(productivity_by_job_stress)

```

```python
# Average Productivity Change by Job Role and Stress Level
# Objective: Understand how job roles and stress levels impact changes in productivity.

plt.figure(figsize=(12, 8))
sns.heatmap(productivity_by_job_stress, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Average Productivity Change by Job Role and Stress Level")
plt.xlabel("Stress Level")
plt.ylabel("Job Role")
plt.show()
```

```python
## Sleep Quality by Work Location
```

```python
# Sleep Quality by Work Location
# Objective: Compare the quality of sleep among employees based on their work location.

sleep_quality_by_location = remote_work.groupby('Work_Location')['Sleep_Quality'].value_counts(normalize=True).unstack()

sleep_quality_by_location.plot(kind='bar',stacked=True,figsize=(10,6),color=['#f4a261', '#2a9d8f', '#e76f51'])
plt.title("Sleep Quality by Work Location")
plt.xlabel("Work Location")
plt.ylabel("Proportion")
plt.legend(title='Sleep Quality')
plt.show
```

```python
# Proportion of Mental Health Conditions by Industry
# Objective: Explore the distribution of mental health conditions across different industries.

conditions_by_industry_proportion=remote_work.groupby('Industry')['Mental_Health_Condition'].value_counts(normalize=True).unstack()

conditions_by_industry_proportion.plot(kind='bar',stacked=True,figsize=(12,6),color=['#f4a261', '#2a9d8f', '#e76f51', '#264653'])
plt.title("Proportion of Mental Health Conditions by Industry")
plt.xlabel("Industry")
plt.ylabel("Proportion")
plt.legend(title='Mental Health Condition')
plt.show()
```

```python
# Physical Activity vs Work-Life Balance
# Objective: Analyze the relationship between physical activity frequency and employees' work-life balance ratings.

activity_balance_crosstab=pd.crosstab(remote_work['Physical_Activity'],remote_work['Work_Life_Balance_Rating'])

sns.heatmap(activity_balance_crosstab,annot=True,cmap='coolwarm',fmt='d')
plt.title("Physical Activity vs Work-Life Balance")
plt.xlabel('Work-Life Balance Rating')
plt.ylabel('Physical Activity')
plt.show
```
